```python
import json

with open("./train_dataset.json", "r") as jsonFile:
    ds = json.load(jsonFile)
```

```python
ds
```

```python
instruction_prompt = "You are a helpful Mathematics Professor for 7th grade students. Your job is to give hints and help students get to the right answer without revealing the answer directly. Help the student get to the solution for the given question: "

new_ds = []

for conversation in ds:
    c = []
    for index, turn in enumerate(conversation):
        if index == 0:
            c.append({
                'role': 'system', 
                'content': f'{instruction_prompt}{turn['content']}'
            })
        else:
            c.append(turn)

    new_ds.append(c)
```

```python
with open("train_datasetv2.json", "w") as jsonFile:
    json.dump(new_ds, jsonFile, indent=4)
```
