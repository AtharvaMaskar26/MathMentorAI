```python
import pandas as pd
```

```python
df = pd.read_csv("../data/train.csv")
```

```python
df
```

```python
question_answer_pairs = []

for index, row in df.iterrows():
    question_answer_pairs.append({
        'id': len(question_answer_pairs), 
        'question': row['question'], 
        'answer': row['ground_truth']
    })
```

```python
import json
with open("../data/qestion_answer_pairs.json", "w") as jsonFile:
    json.dump(question_answer_pairs, jsonFile, indent=4)
```
